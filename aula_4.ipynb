```python
for n1 in [8, 7, 17, 25, 38]:
  print(n1)

seq = [3, 57, 67, 22, 8]
for n2 in seq: #aqui ele "desempacota" a lista e vê os valores nela
  print(n2)

'''
Os dois imprimem a mesma coisa, porém é uma boa prática colocar uma lista em uma variável ao invés de colocar na lista no for.
'''
```

```python
for numero in range(100000000): #range é usado para eu mostrar quantas vezes eu quero iterar literalmente. Nesse caso iteraria esses milhões de vezes.
  print(numero)
  if numero == 9:
    print("Achei o Quatro")
    break #sai do laço
print("Até mais")
```

```python
lista = [12, True, "Teste", 43.6]
len(lista) #pega o tamanho da lista

for i in range(len(lista)): #len(lista) retorna um valor inteiro
  print(i) #ele vai printar a lista que o range cria com os números de 0 ao tamanho len

for i in lista:
  print(i) #imprime os elementos da lista
```

```python
for i in range(1,20,5): # o 1,20 e 5 não é uma lista e sim argumentos da função range(inicio, fim, salto)
  print(i)

lista = [12, True, "Teste", 43.6]

for i in range(len(lista)):
  print(lista[i]) #puxando o índice da lista com base no "i"
```

```python
for x in [1, 10, 20, 30, 40 ,50]:
  if x == 30:
    continue # é o oposto do break, ele não para o laço ele pede para continuar e retira o valor pedido
  print(x)
```

```python
inicio = int(input("Informe o primeiro número:"))
fim = int(input("Informe o número final:"))
salto = int(input("Informe o salto:"))
texto = "Cálculo:"
soma = 0
for numero in range(inicio, fim, salto):
  soma = soma + numero
  texto = texto + str(numero)
  if numero > 50:
    texto = texto + "\n Passou de 50"
    break
  if numero != fim-1:
    texto = texto + "+"
print(f"{texto}")
print(f"Soma: {soma}")
```
