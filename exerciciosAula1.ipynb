```python
firstText = input("Escreva sua primeira frase:")
secondText = input("Escreva sua segunda frase:")

print("O usuário digitou:", firstText, " e ", secondText)

print("O usuário digitou: " + firstText + " e " + secondText) #Este formato se chama concatenação de strings, não funciona com outros tipos de variáveis

print(f"O usuário digitou {firstText} e {secondText}") #Este "f" é de format

print("O usuário digitou: %s e %s" %(firstText,secondText)) # %s é para strings
```

```python
widthRec = float(input("Digite a largura do retângulo: "))
heightRec = float(input("Digite a altura do retângulo: "))

areaRec = widthRec * heightRec
perimeterRec = 2*(widthRec + heightRec)

print("O perimetro do retângulo é:", perimeterRec)
print("A área do retângulo é: %f" %(areaRec)) # %f pois é uma float
```

```python
price = float(input("O valor total do produto é: "))
priceTax = float(input("A taxa aplicada a ele é: "))
periodMonths = float(input("O tempo de atraso é: "))

# % em programação significa "módulo" que é 0 ou um por exemplo, 4%2 = 0 pois o resto da divisão é zero

# 3 // 2, esse "//" ele indica que eu quero apenas o resultado inteiro da divisão

prestacaoPrice = (price + (price*(priceTax/100)* periodMonths))
print("O valor da prestação é:", prestacaoPrice)
```
