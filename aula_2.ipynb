```python
import numpy as np

nomeAluno = input("Nome do aluno: ")
notaP1 = float(input("Nota na P1: "))
notaP2 = float(input("Nota na P2: "))
notaP3 = float(input("Nota na P3: "))
notas = np.array([notaP1, notaP2, notaP3])

mediaNota = np.sum(notas)/notas.size
mediaDireta = np.mean(notas)

mediaArrend = round(mediaNota,2) #arredondamento, round(valor, numero de casa decimais)

print("A nota não arredondada é %f" %(mediaNota))
print(f"A media do Aluno, {nomeAluno}, foi de {mediaArrend} e a média da turma foi {mediaArrend}")
```

```python
import math as mt
import numpy as np

catetoA = float(input("Tamanho do cateto A: "))
catetoB = float(input("Tamanho do cateto B: "))
somaQuadratica = mt.pow(catetoA,2) + mt.pow(catetoB,2)
pitagorasHipo = mt.pow(somaQuadratica,0.5)

pitagorasDireto = mt.pow((catetoA**2 + catetoB**2),0.5)
print(pitagorasDireto)

pitagorasOutro = mt.sqrt(catetoA**2 + catetoB**2)
print(pitagorasOutro)

print("A hipotenusa do triângulo em questão é de", pitagorasHipo)
```
