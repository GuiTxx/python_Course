```python
import datetime as dt

ano_nascimento = int(input("Digite o ano do nascimento: "))
mes_nascimento = int(input("Digite o mês do nascimento: "))
dia_nascimento = int(input("Digite o dia do nascimento: "))

hoje = dt.date.today()
ano_atual = hoje.year
mes_atual = hoje.month
dia_atual = hoje.day

idade = ano_atual - ano_nascimento
if mes_atual < mes_nascimento:
  idade -= 1 # isso significa que "idade = idade - 1"
elif mes_atual == mes_nascimento:
    #se o mês for igual, já verifica o dia
    if dia_atual < dia_nascimento:
      idade -= 1

print("A sua idade é:", idade)

if idade >= 18:
  print("Você é maior de idade.")
else:
  print("Você é menor de idade!")
```

```python
'''
Faça o usuário digitar dois números inteiros e dizer se estes números são iguais ou diferentes
'''
num_1 = float(input("Digite um número aleatório:"))
num_2 = float(input("Digite um outro número aleatório:"))

if num_1 == num_2:
  print("Os números são iguais.")
else:
  dif = num_1 - num_2
  print(f"Os números são diferentes e possuem diferença de {dif}")
```
