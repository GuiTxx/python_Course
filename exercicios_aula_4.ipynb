```python
#Exercício 2

lista = []
for i in range(10):
  valor = float(input(f"Digite o {i+1}º valor:"))
  lista.append(valor)
print(lista)

maior_valor = 0
menor_valor = lista[0]
for j in range(10):
  numero = lista[j]
  if numero > maior_valor:
    maior_valor = numero
  elif numero < menor_valor:
    menor_valor = numero
  else: continue

media = sum(lista)/ len(lista)

print(maior_valor)
print(menor_valor)
print(media)
```

```python
#Exercício 3

numero_clientes = int(input("Quantidade de clientes:"))
estado_cliente = []

for i in range(numero_clientes):
  temperatura = float(input(f"Temperatura do {i+1}º Cliente:\n"))
  estado_cliente.append(temperatura)

for j in range(numero_clientes):

  temp_referencia = estado_cliente[j] #atribuo como referência cada temperatura da lista de temperaturas

  if temp_referencia < 37.2:
    print(f"O {j+1}º cliente apresenta estado normal.\n")

  elif 37.2 <= temp_referencia < 38.0:
    print(f"O {j+1}º cliente apresenta estado febril.\n")

  elif 38.0 <= temp_referencia <= 39.0:
    print(f"O {j+1}º cliente apresenta estado de febre corporal.\n")

  elif temp_referencia > 39.0:
    print(f"O {j+1}º cliente apresenta estado de febre corporal alta!\n")

  else:
    print("Erro de leitura dos dados.")

print("A quantidade de clientes analisados foi de:", numero_clientes)

media = sum(estado_cliente)/len(estado_cliente)
print("A temperatura média dos clientes foi de:",media,"°C")

```
